```python
#Nupy project
#Tic-Tac -TOE pROJECTY
#WE CAN AL;SO MAKE THIS PROJETC USING LIST IN PYHTON
# BUT WE WILL SEE USING NUPY
import numpy as np
board=np.zeros((3,3),dtype=int)
#intially board have 0
#and 1 will represet x and -1 will represnt o
```

```python
def print_board(b):
  symbols={0:" ",1:"x",-1:"0"}
  for r in range(3):
    row=" | ".join(symbols[val] for val in b[r])
    print(" "+row)
    if r<2 :
      print("---+---+---")
  print()
print_board(board)
```

```python
def check_winner(b):
  if 3 in np.sum(b,axis=1) or 3 in np.sum(b,axis=0):
    return 'x'
  if -3 in np.sum(b,axis=1) or -3 in np.sum(b,axis=0):
    return 'O'
  if np.trace(b)==3 or np.trace(np.fliplr(b)) == 3:
    return 'x'
  if np.trace(b)==-3 or np.trace(np.fliplr(b))==-3: #TRACE OF FLIPPING ARRAY
    return 'O'
  if not 0 in b: # ek bhi 0 nahi bacha hua hai
    return "DRAW"
  return None
#print_board(board)
```
